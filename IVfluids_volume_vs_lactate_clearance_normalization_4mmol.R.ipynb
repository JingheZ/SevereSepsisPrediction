```r
# select the patients with at least one lactate > 4mmol/L
data_df2.2 <- data_df2.1[data_df2.1$id %in% data_df2.ptid,]
data_df2.2 <- data_df2.2[order(data_df2.2$id, data_df2.2$charttime),]
```

```r
#get the time of severe sepsis for each patient

sepsis.time <- function(data) {
  event.time <- NA
  clear.time <- NA
  normalize.time <- NA
  lastlactate.time <- data$charttime[nrow(data)]
  ind <- NA
  if (nrow(data) < 1) {
    times <- c(event.time, clear.time, normalize.time, lastlactate.time)
    return(times)   
  } 
  else {
    event.time <- data$charttime[data$valuenum > 4][1]
    initial.lactate <- data$valuenum[data$valuenum > 4][1]
    ind <- which(data$valuenum > 4)[1]
    if (!is.na(ind) &(ind < nrow(data))) {
      data1 <- data[(ind:nrow(data)),]
      clear.time <- data1$charttime[(data1$valuenum <= initial.lactate*0.9)][1]
      normalize.time <- data1$charttime[data1$valuenum <= 2.5][1]
    }
#    init <- c(initial.lactate)
    times <- c(event.time, clear.time, normalize.time, lastlactate.time)
    return(times)   
  }
}


df2_times <- ddply(data_df2.2, .(id), sepsis.time)
names(df2_times) <- c('id', 'event.time', 'clear.time', 'normalize.time', 'lastlactate.time')

df2_times[1:5,]

```

```r
str(fluids3)
saline_volume <- ddply(fluids4, .(id), summarize, vol=sum(volume))
head(saline_volume)
dim(saline_volume)
```

```r
# #add the event time, clear time and normalization time, last lactate time to the aggregagted normal saline volume data
sepsis.time$clear.dftime <- difftime(sepsis.time$clear.time, sepsis.time$event.time, unit='hour')
sepsis.time$normalize.dftime <- difftime(sepsis.time$normalize.time, sepsis.time$event.time, unit='hour')
sepsis.time$lastlactate.dftime <- difftime(sepsis.time$lastlactate.time, sepsis.time$event.time, unit='hour')

head(sepsis.time)

saline_data <- merge(saline_volume, sepsis.time, by.x = 'id', by.y = 'id')
head(saline_data)
dim(saline_data)
```

```r
library(ggplot2)
print('density plot of the distribution of fluid volume (0~6 hours)')
ggplot(saline_data, aes(x=vol)) + 
  geom_histogram(aes(y=..density..),      # Histogram with density instead of count on y-axis
                 binwidth=10,
                 colour="black", fill="white") +
  geom_density(alpha=.2, fill="#FF6666")
```

```r

#create dataframe for patients with fluids within 0~6 hours, censor = 1; else 0
saline_data2 <- saline_data[!is.na(saline_data$event.time),]
saline_data2$censor <- NA
saline_data2$time <- NA
for (i in 1:nrow(saline_data2)) {
  if (is.na(saline_data2$clear.dftime[i])) {
    saline_data2$time[i] <- saline_data2$lastlactate.dftime[i]
    saline_data2$censor[i] <- 0
  }  else {
    saline_data2$time[i] <- saline_data2$clear.dftime[i]
    saline_data2$censor[i] <- 1   
  }
}
head(saline_data2)
```

```r
ggplot(saline_data2, aes(x=time)) +
  geom_histogram(binwidth=.5, colour="black", fill="white") +ggtitle('Histogram of lactate clearance time')
#remove outliers 
summary(saline_data2$time)
x <- quantile(saline_data2$time, c(0.01, 0.99), )
saline_data2a <- saline_data2[(saline_data2$time > x[1]) & (saline_data2$time < x[2]), ]
x
```

```r
summary(saline_data2a$vol)
```

```r
#survival analysis; Kaplan-Merier Curves
library(survival)

volumes <- cut(saline_data2a$vol, c(0, 300, 1000, 5000, 15000), right = F)
survfluid.time <- survfit(Surv(time, censor) ~ volumes, data = saline_data2a)
print(survfluid.time)
plot(survfluid.time, lty=c(1,1,5,6), col = c('red', 'green', 'black', 'blue'), xlab="Time (0~120 hrs)", ylab="Lactate Clearance Probability", fun = 'event')
legend(80, 0.3, c('[0,300)', '[300, 1000)', '[1000,5000)', '[5000,15000)'), lty=c(1,1,5,6), col = c('red', 'green', 'black', 'blue')) 
```
